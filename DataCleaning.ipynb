```python
import pandas as pd 
import numpy as np

df = pd.read_csv('data.csv')
```

```python
df['confirmed_cases'].fillna(df['confirmed_cases'].mean(), inplace=True)
df['deaths'].fillna(df['deaths'].mean(), inplace=True)
df['recovered'].fillna(df['recovered'].mean(), inplace=True)
df['tests_done'].fillna(df['tests_done'].mean(), inplace=True)
df['vaccinations'].fillna(df['vaccinations'].mean(), inplace=True)
df.drop_duplicates(inplace=True)
df['date'] = pd.to_datetime(df['date'])


df.info()
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
